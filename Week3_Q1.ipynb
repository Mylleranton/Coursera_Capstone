```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import bs4 as bs
from bs4 import BeautifulSoup
```

```python
# Lets define the url and get data using beautifulsoup
import requests
url=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(url, 'lxml')

# Table is in .table
table = soup.table

# Make dataframe and set data types
df = pd.read_html(str(table))[0]
df = df.astype(str)
df.shape
```

```python
# Drop all rows with N/A Borough
df = df[~df["Borough"].isin(["Not assigned"])]
df.shape
```

```python
# Ensure that Postal codes with multiple Neighbourhoods have them separated by
# commas (UPDATE: Wiki table is updated to separate with commas instead of slashes)
# OBSOLETE CODE:
df['Neighborhood'] = df['Neighborhood'].str.replace(' /',',')
df.shape
```

```python
# If a cell has a borough but a Not assigned neighborhood, 
# then the neighborhood will be the same as the borough

df.loc[df["Neighborhood"] == "NaN", "Neighborhood"] = df["Borough"]
df.shape
```
